**Q1. For the year 2015, which were the top 10 countries that used the most non-renewable energy source and how did it comapre with renewable energy for this countries?**

The scope of our analysis is for the period from 2005 to 2015.

```python
electricity_df_q1=electricity_df[electricity_df['Year']==2015]
#To summarise non-renewable we create a seperate column.
electricity_df['Total Non Renewable(GWh)']=electricity_df['Electricity from coal (TWh)']+electricity_df['Electricity from gas (TWh)']+electricity_df['Electricity from oil (TWh)']+electricity_df['Electricity from nuclear (TWh)']
#convert 'Total non-renewable in Gigawatt hour and round off
electricity_df['Total Non Renewable(GWh)']=round(electricity_df['Total Non Renewable(GWh)']*100000,0)
#Repeat same steps for renewable
#To summarise  renewable we create a seperate column.
electricity_df['Total Renewable(GWh)']=electricity_df['Electricity from hydro (TWh)']+electricity_df['Electricity from other renewables (TWh)']+electricity_df['Electricity from solar (TWh)']+electricity_df['Electricity from wind (TWh)']
#convert 'Total renewable in Gigawatt hour and round off
electricity_df['Total Renewable(GWh)']=round(electricity_df['Total Renewable(GWh)']*100000,0)
# Display the data
electricity_df
```

Our focusing to find the countries using non-renewable in 2015. limit our data to 2015

```python
# Including only 2015 data. Selecting only the 2015 data
electricity_df_q1=electricity_df[electricity_df['Year']==2015]

# Calculating the sum of Total Non reneweable by suming the columns
electricity_df_q1=electricity_df_q1.groupby('Country Name').sum().sort_values(by='Total Non Renewable(GWh)',ascending=True).tail(10)
# Ploting the data
electricity_df_q1.plot(
   kind='barh',# Horizontal graph
   y=['Total Non Renewable(GWh)','Total Renewable(GWh)'],# y-axis , two columnns passed as a list
   title='Top 10 Countries (Highest Use of Non Renewable & Use oF Renewable)',#Specify title
   stacked=True # Stacking columns for renewable and non-renewable
)
#Display the data # Sorting in table is in descending order to match the readability of the plot.
electricity_df_q1[['Total Non Renewable(GWh)','Total Renewable(GWh)']].sort_values(by='Total Non Renewable(GWh)',ascending=False)
```

**Answer**

Form the above figure we can see that China is country with uses the highest Non-Renewable source of energy. Also, among the top 10 of Non Renewable it is also the uses most Renewable energy. Other countries in the top10 countries using non-renewable include United States, India, Russia, Japan, South Korea, France, Germany, Saudi Arabia and United Kingdom.
